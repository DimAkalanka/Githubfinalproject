```python
import pandas as pd
import numpy as np
```

```python
all_data = pd.read_csv('/content/sample_data/daily_data.csv')
all_data
#
```

# **Conversion of Catogorical to Numerical**

```python
from sklearn.preprocessing import LabelEncoder
label_encode = LabelEncoder()
all_data['city_id'] = label_encode.fit_transform(all_data['city_id'])
```

# Conversion of Time Values to numbers

```python
all_data['sunrise'] = pd.to_datetime(all_data['sunrise'], format='%I:%M %p')
```

```python
all_data['sunset'] = pd.to_datetime(all_data['sunset'], format='%I:%M %p')
```

```python
reference_datetime = pd.to_datetime("1900-01-01 00:00:00")
all_data['sunrise'] = (all_data['sunrise'] - reference_datetime).dt.total_seconds()
```

```python
reference_datetime1 = pd.to_datetime("1900-01-01 12:00:00")
all_data['sunset'] = (all_data['sunset'] - reference_datetime1).dt.total_seconds()
```

```python
all_data
```

# Seperate data to nulls and non- nulls

```python

df_nonull = all_data.dropna(how='any')
dropped_df = all_data[~all_data.index.isin(df_nonull.index)]
```

## non_nulls according to dependent variable

```python
df_nonull
```

## NUll values according to dependent variable

```python
dropped_df
```

## Drop unwanted columns and assign Independent and dependent variables

```python
x = df_nonull.drop(['day_id','condition_text'], axis=1)
y = df_nonull['condition_text']
```

# Split data to train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

# Train a model using Randomforestclassifier

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=150,max_depth = 200, random_state=1)
model.fit(x_train, y_train)
```

# Do a Prediction

```python
y_hat = model.predict(x_test)
```

# Evaluation of Model

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_test, y_hat,average='weighted')
recall = recall_score(y_test, y_hat,average='weighted')
f1score = f1_score(y_test, y_hat,average='weighted')

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1score:.2f}")
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_hat)
print(f"Accuracy: {accuracy:.2f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=15)
knn.fit(x_train, y_train)
```
